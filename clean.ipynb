```python
import pandas as pd

billings_df = pd.read_csv("C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\billings.csv")
calls_df = pd.read_csv("C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\cc_calls.csv")
emails_df = pd.read_csv("C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\emails.csv")
renewal_calls_df = pd.read_csv("C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\renewal_calls.csv")
```

```python
import pandas as pd

# ✅ Optional: Clean column names
for df in [billings_df, calls_df, emails_df, renewal_calls_df]:
    df.columns = df.columns.str.strip()

# ✅ Step 1: Filter emails for Time_to_Renewal = "14_out"
emails_14_out_df = emails_df[emails_df["Time_to_Renewal"] == "14_out"]

# ✅ Step 2: Extract Co_Ref list
co_ref_14_out = emails_14_out_df["Co_Ref"].unique()

# ✅ Step 3: Filter other datasets using Co_Ref
billings_filtered_df = billings_df[billings_df["Co_Ref"].isin(co_ref_14_out)]
calls_filtered_df = calls_df[calls_df["Co_Ref"].isin(co_ref_14_out)]
renewal_calls_filtered_df = renewal_calls_df[renewal_calls_df["Co_Ref"].isin(co_ref_14_out)]

# ✅ Step 4: Save each filtered dataframe to CSV
emails_14_out_df.to_csv(
    "C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\emails_14_out.csv",
    index=False
)

billings_filtered_df.to_csv(
    "C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\billings_14_out.csv",
    index=False
)

calls_filtered_df.to_csv(
    "C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\calls_14_out.csv",
    index=False
)

renewal_calls_filtered_df.to_csv(
    "C:\\Users\\MadanRajMA\\Documents\\DS\\Dataset_new\\renewal_calls_14_out.csv",
    index=False
)
```
