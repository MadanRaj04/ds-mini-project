```python
import pandas as pd

billings = pd.read_csv('../../data/cleaned/billings_cleaned.csv', low_memory=False) 
emails = pd.read_csv('../../data/cleaned/emails_cleaned.csv', low_memory=False) 
renewal_calls = pd.read_csv('../../data/cleaned/renewal_calls_cleaned.csv', low_memory=False) 
cc_calls = pd.read_csv('../../data/cleaned/cc_calls_cleaned.csv', low_memory=False)
```

```python
merged_dataset = emails.merge(billings, on='co_ref', how='left').merge(renewal_calls, on='co_ref', how='left').merge(cc_calls, on='co_ref', how='left')
```

```python
merged_dataset.head()
```

```python
import os
os.makedirs('../../data/merged', exist_ok=True)
```

```python
merged_dataset.shape
```

```python
merged_dataset['call_date_x'] = pd.to_datetime(merged_dataset['call_date_x'], errors='coerce')
merged_dataset['prospect_renewal_date'] = pd.to_datetime(merged_dataset['prospect_renewal_date'], errors='coerce')

filtered_dataset = merged_dataset[
    (merged_dataset['call_date_x'] >= merged_dataset['prospect_renewal_date'] - pd.Timedelta(days=45)) &
    (merged_dataset['call_date_x'] <= merged_dataset['prospect_renewal_date'] - pd.Timedelta(days=14))
]
```

```python
filtered_dataset.shape
```

```python
filtered_dataset.to_csv('../../data/merged/merged.csv', index=False)
```
