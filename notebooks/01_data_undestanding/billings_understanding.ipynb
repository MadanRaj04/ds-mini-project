```python
null_summary = pd.DataFrame({
    'dtype'     : billings.dtypes,
    'null_count': billings.isnull().sum(),
    'null_pct'  : (billings.isnull().sum() / len(billings) * 100).round(1)
}).sort_values('null_pct', ascending=False)

print(null_summary.to_string())
```
