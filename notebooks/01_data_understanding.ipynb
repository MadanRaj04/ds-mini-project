## Step 1: Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load all 4 raw datasets
renewal  = pd.read_csv('../data/raw/renewal_calls.csv', low_memory=False)
billing  = pd.read_csv('../data/raw/billings.csv', low_memory=False)
cc_calls = pd.read_csv('../data/raw/cc_calls.csv', low_memory=False)
emails   = pd.read_csv('../data/raw/emails.csv', low_memory=False)

print('All datasets loaded successfully!')
```

## Step 2: Quick Overview of All Datasets

```python
# Print shape and basic info for all datasets
datasets = {
    'Renewal Calls': renewal,
    'Billings': billing,
    'CC Calls': cc_calls,
    'Emails': emails
}

for name, df in datasets.items():
    print(f"\n{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
    print(f"Duplicate Rows: {df.duplicated().sum()}")
```

```python
# Unique values per column
for col in renewal.columns:
    print(f"{col}: {renewal[col].nunique()} unique values")
    if renewal[col].nunique() < 15:  # Show value counts for low-cardinality columns
        print(renewal[col].value_counts())
    print()
```

## Step 7: How Datasets Connect (Join Key Analysis)

```python
# Check Co_Ref across all datasets
print('Unique Customers per Dataset:')
print(f"  Renewal Calls: {renewal['Co_Ref'].nunique():,} customers")

# Check if Co_Ref exists in other datasets
for name, df in [('Billings', billing), ('CC Calls', cc_calls), ('Emails', emails)]:
    if 'Co_Ref' in df.columns:
        print(f"  {name}: {df['Co_Ref'].nunique():,} customers")
    else:
        # Try to find similar column
        potential = [c for c in df.columns if 'ref' in c.lower() or 'co' in c.lower() or 'customer' in c.lower() or 'id' in c.lower()]
        print(f"  {name}: 'Co_Ref' NOT found. Potential join columns: {potential}")
```

```python
# Find overlapping customers between datasets
renewal_customers = set(renewal['Co_Ref'].dropna().unique())

for name, df in [('Billings', billing), ('CC Calls', cc_calls), ('Emails', emails)]:
    if 'Co_Ref' in df.columns:
        other_customers = set(df['Co_Ref'].dropna().unique())
        overlap = renewal_customers & other_customers
        print(f"Renewal ∩ {name}: {len(overlap):,} common customers")
    else:
        print(f"{name}: Cannot check overlap — Co_Ref column not found")
```
